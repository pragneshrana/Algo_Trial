```
import matplotlib.pyplot as plt
import numpy as np
X = np.random.random((10,1))
print(X)
```

```
plt.plot(X,'*-')
plt.show()
```

```
x = np.linspace(0,10,100) #generate 100 numbrs
y = np.power(x,0.5)
plt.plot(x,y,'*-')
plt.show()
```

```
import seaborn as sns
```

```
sns.set() #to set style
sns.lineplot(x,y)
```

```
import pandas as pd
data = pd.read_csv('happiness.csv')
data['Dummy'] = np.random.random((data.shape[0],1))
```

```
data.head()
```

```
ax = sns.scatterplot(x='Mean',y='N=',data=data)
```

```
ax = sns.scatterplot(x='Mean',y='N=',hue='Dummy',data=data)
```

```
ax = sns.distplot(data['Mean'],kde=True,rug=True,bins=50)
```

```
ax = sns.boxplot(y='Mean',x='Gender',data = data)
```

```
uniform_data = np.random.rand(10,12)
print(uniform_data)
ax = sns.heatmap(uniform_data)
```

```
ax = sns.heatmap(uniform_data,cmap='BuPu_r')
```
